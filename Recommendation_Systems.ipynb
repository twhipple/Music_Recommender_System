# Libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data

```python
# Look at artists data
df = pd.read_csv("Data/data_by_artist.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# lets see top10 artists who sang more songs than others
top10 = pd.DataFrame({'No of songs':df['artists'].value_counts().head(10)})
top10.plot.bar()
plt.title('Top 10 artists')
plt.xlabel('Artists')
plt.ylabel('No of song')
plt.show()
```

```python
# Look at the damage_grades - looks pretty unbalanced
top_artist = df['artists'].value_counts().head(10)
plt.figure(figsize=(16,5))
sns.barplot(y=top_artist.index, x=top_artist.values, palette='CMRmap')
plt.title('Top Artists', fontsize=18)
plt.ylabel('Artists', fontsize=12)
plt.xlabel('Number of Songs', fontsize=12)
plt.legend(labels=["1 Low Damage","2 Medium Damage","3 Complete Destruction"])
plt.show()
```

```python
# finding correlation
corr = df.corr()
# visualizing correlaiton with heatmap
plt.figure(figsize=(20,8))
sns.heatmap(corr, vmax=1, vmin=-1, center=0,linewidth=.5,square=True, 
            annot = True, annot_kws = {'size':8},fmt='.1f', cmap='CMRmap')
plt.title('Correlation')
plt.show()
```
